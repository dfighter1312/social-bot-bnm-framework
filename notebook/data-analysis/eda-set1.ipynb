### Engineered Features
These includes:
1. User features
   - [x] account_age
   - [x] screen_name_length
   - [x] Num_digits_in_screen_name
   - [x] name_length
   - [x] num_digits_in_name
   - [x] description_length
   - [x] screen_name_likelihood
2. Activity features
   - [x] followers_growth_rate
   - [x] friends_growth_rate
   - [x] favorites_growth_rate
   - [x] listed_growth_rate
   - [x] account_reputation
3. Tweet features: Not available in this set
4. Network features: We don't use this

```python
from nltk.lm.preprocessing import padded_everygram_pipeline, pad_both_ends
from nltk.util import trigrams
from nltk.lm import MLE

# helper functions
def get_screen_name_likelihood(series):
    sequence = series.apply(lambda x: list(x.lower())).values.tolist()
    return get_likelihood_array(sequence)
    
def get_likelihood_array(sequence, n=3):
    train_data, padded_sent = padded_everygram_pipeline(n, sequence)
    mle = MLE(n)
    mle.fit(train_data, padded_sent)
    
    s = np.zeros((len(sequence),))
    for i, name in enumerate(sequence):
        tri = trigrams(pad_both_ends(name, n=3))
        total_score = 1
        count = 0
        for ele in tri:
            score = mle.score(ele[2], [ele[0], ele[1]])
            total_score *= score
            count += 1
        s[i] = total_score ** (1/count)
    return s
```

```python

def extracted_user_features(df):
    df_return = pd.DataFrame()
    df_return['screen_name_length'] = df['screen_name'].str.len()
    df_return['num_digits_in_screen_name'] = df['screen_name'].str.count('\d')
    df_return['name_length'] = df['name'].str.len()
    df_return['num_digits_in_name'] = df['name'].str.count('\d')
    df_return['description_length'] = df['description'].str.len()
    df_return['screen_name_likelihood'] = get_screen_name_likelihood(df['screen_name'])
    return df_return

def extracted_activity_features(df):
    # NOTE: age is extracted here for convenient
    df_return = pd.DataFrame()
    age = df_return['age'] = (
        (
            pd.to_datetime(df['updated'])
            - pd.to_datetime(df['created_at']).dt.tz_localize(None)
        ) 
        / np.timedelta64(1, 'Y')
    )
    df_return['followers_growth_rate'] = df['followers_count'] / age
    df_return['friends_growth_rate'] = df['friends_count'] / age
    df_return['favourites_growth_rate'] = df['favourites_count'] / age
    df_return['listed_growth_rate'] = df['listed_count'] / age
    df_return['reputation'] = df['followers_count'] / (df['followers_count'] + df['friends_count'])
    return df_return
    
```

```python
''' NOTE:
    The selected features are basically all the original features that is
    usable (in the sense that it's numeric or bool)
    the not chosen ones either already got extracted into sth else
    or is not usable (like profile_background_image_url_https) for example
'''
SELECTED_FEATURES = [
    'statuses_count', 'followers_count', 'friends_count', 'favourites_count',
    'listed_count', 'default_profile', 'default_profile_image', 'geo_enabled',
    'profile_use_background_image', 'profile_background_tile', 'utc_offset',
    'is_translator', 'protected', 'verified'
]

def feature_engineering(df, label):
    df_new = pd.concat(
        [
            df[SELECTED_FEATURES],
            extracted_user_features(df),
            extracted_activity_features(df),
        ], axis = 1
    )
    df_new['label'] = label
    return df_new
```

```python
corr = df.corr()# plot the heatmap
figure = plt.figure(figsize=(30,30))

mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, k = 1)] = True

sns.heatmap(
    corr,
    mask=mask,
    robust= True, 
    xticklabels=corr.columns, 
    yticklabels=corr.columns, 
    annot=True, 
    cmap=sns.diverging_palette(220, 20, as_cmap=True)
)
figure.patch.set_facecolor('white')
plt.show()
```

### Observation

```python
high_corr = np.abs(corr) > 0.5
matching_indexes = np.where(high_corr)

high_correlated_pairs = set()
for x,y in zip(matching_indexes[0],matching_indexes[1]):
    if x==y: continue
    high_correlated_pairs.add(frozenset({corr.index[x], corr.columns[y]}))

high_correlated_pairs = np.array(list(high_correlated_pairs))
for pair in np.unique(high_correlated_pairs):
    pair = tuple(pair)
    print(f'{pair} - cor: {corr.at[pair[0], pair[1]]}')
```

### Observation
Here are the most significant correlations (>0.7):
1. Growth rate and their counter part
   - ('listed_count', 'listed_growth_rate') - cor: 0.976904342142999
   - ('favourites_count', 'favourites_growth_rate') - cor: 0.870932227862492
   - ('friends_growth_rate', 'friends_count') - cor: 0.9309833295476694
   - ('followers_growth_rate', 'followers_count') - cor: 0.9544926035681912

### Bot vs Man on Numbers
follower, friend, list, faavorites

```python
df = df.reset_index(drop=True)
```

```python
import math
feature_list = [
    'friends_count', 'followers_count', 'listed_count', 'favourites_count',
    # 'friends_growth_rate', 'followers_growth_rate', 'listed_growth_rate', 'favourites_growth_rate',
]

n_cols = min(3, len(feature_list))
n_rows = math.ceil(len(feature_list)/float(n_cols))
figure = plt.figure(figsize=(n_cols*8, n_rows*5))


for i, feature in enumerate(feature_list):
    ax = figure.add_subplot(n_rows, n_cols, i+1)
    sns.barplot(data=df, x='label', y=feature, ax=ax)
    ax.set_title(feature)
figure.patch.set_facecolor('white')

# mid = (figure.subplotpars.right + figure.subplotpars.left)/2
plt.suptitle('Bot vs Human Barcharts')
plt.show()
```

Genuine account have way more favor, listed and follower in sum

```python
''' NOTE
    Histplot looks very ugly since data is high value and not normalized
    So to use with log scale, value >0 need to be selected
'''
for i, feature in enumerate(feature_list):
    figure = plt.figure(figsize=(10,10))

    test = df[df[feature] > 0.0]
    sns.histplot(data=test, x=feature, hue='label', log_scale=True)

    figure.patch.set_facecolor('white')
    plt.suptitle(f'Histogram {feature} bot vs human (log_scaled)')
    plt.show()
```

**Observations**
- Human's friend, follower and favorites tend to follow normal distribution

- As for listed, both have a tendency to not be listed in many!
